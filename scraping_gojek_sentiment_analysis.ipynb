```python
import pandas as pd
import time
from google_play_scraper import Sort, reviews
```

```python
app_id = 'com.gojek.app'

results, _ = reviews(
    app_id,
    lang='id',
    country='id',
    sort=Sort.NEWEST,
    count=500000,
)
```

```python
columns = [list(result.keys()) for result in results[:1]]
print(columns)
```

```python
values = [list(result.values()) for result in results]

df = pd.DataFrame(values, columns=columns[0])
df.head()
```

```python
df.to_csv("gojek_sentiment_analysis.csv", index=False)
```
